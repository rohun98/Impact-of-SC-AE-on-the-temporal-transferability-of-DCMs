# Primary Education Enrolment

```python
#Reading Files

file_path_18 = "Primary/01-leerlingen-po-soort-po-cluster-leeftijd-2018-2019 (1).csv"
file_path_22 = "Primary/01.-leerlingen-po-soort-po,-cluster,-leeftijd-2022-2023.csv"

# Read the CSV file into a DataFrame
ps_data_18 = pd.read_csv(file_path_18, sep=';', encoding='latin-1')
ps_data_22 = pd.read_csv(file_path_22, sep=';', encoding='latin-1')

ps_data_18.head(5)
```

```python
# Making a copy with 2018 as reference data

ps =  ps_data_18.copy()
ps_22 = ps_data_22.copy()

# Keeping only relevant columns
ps =    ps   [['PEILDATUM', 'POSTCODE_VESTIGING', 'GEMEENTENUMMER', 'GEMEENTENAAM', 'TOTAAL']]
ps_22 = ps_22[['PEILDATUM', 'POSTCODE_VESTIGING', 'GEMEENTENUMMER', 'GEMEENTENAAM', 'TOTAAL']]
```

```python
# Converting PC6 to PC4 digit and then aggregating it on a PC4 level

ps['POSTCODE_VESTIGING'] = ps['POSTCODE_VESTIGING'].astype(str).str[:4]
ps_22['POSTCODE_VESTIGING'] = ps_22['POSTCODE_VESTIGING'].astype(str).str[:4]

ps_pc4_agg_2018 = ps.groupby('POSTCODE_VESTIGING')['TOTAAL'].sum().reset_index()
ps_pc4_agg_2022 = ps_22.groupby('POSTCODE_VESTIGING')['TOTAAL'].sum().reset_index()
```

```python
#Ensuring consistency in Gemeente Name and number for each PC4 code

ps_pc4_agg_2018 = ps.groupby('POSTCODE_VESTIGING').agg({
    'TOTAAL': 'sum',
    'GEMEENTENUMMER': 'first',
    'GEMEENTENAAM': 'first'
    }).reset_index()



ps_pc4_agg_2022 = ps_22.groupby('POSTCODE_VESTIGING').agg({
    'TOTAAL': 'sum',
    'GEMEENTENUMMER': 'first',
    'GEMEENTENAAM': 'first'
    }).reset_index()
```

```python
# Replacing blank columns with NA and then dropping postal code NAN values




ps_pc4_agg_2018['POSTCODE_VESTIGING'] = ps_pc4_agg_2018['POSTCODE_VESTIGING'].replace(' ', pd.NA)
ps_pc4_agg_2018['GEMEENTENUMMER'] = ps_pc4_agg_2018['GEMEENTENUMMER'].replace(' ', pd.NA)
ps_pc4_agg_2018['GEMEENTENAAM'] = ps_pc4_agg_2018['GEMEENTENAAM'].replace(' ', pd.NA)
ps_pc4_agg_2018 = ps_pc4_agg_2018.dropna(subset=['POSTCODE_VESTIGING'])





ps_pc4_agg_2022['POSTCODE_VESTIGING'] = ps_pc4_agg_2022['POSTCODE_VESTIGING'].replace(' ', pd.NA)
ps_pc4_agg_2022['GEMEENTENUMMER'] = ps_pc4_agg_2022['GEMEENTENUMMER'].replace(' ', pd.NA)
ps_pc4_agg_2022['GEMEENTENAAM'] = ps_pc4_agg_2022['GEMEENTENAAM'].replace(' ', pd.NA)
ps_pc4_agg_2022 = ps_pc4_agg_2022.dropna(subset=['POSTCODE_VESTIGING'])
```
